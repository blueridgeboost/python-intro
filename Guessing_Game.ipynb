The Python code that declares and sets a variable named ```number``` to a random number between 1 and 100 (inclusive) is shown below.

````python
# We have to tell Python we use the random library
import random

number = random.randint(1, 101)
````

The ```forever``` block is written in Python as ```while(True):```

The ```ask``` block does several actions.
1. Prints a bubble with the message
2. Displays a box for the user to enter text
3. Stores the value introduced by the user in a special variable called ```answer```.


In Python, we achive the same by using the ```input``` function.

````python
answer = int(input("What number am I thinking of?"))
````

````python
if (number == answer):
  print("Well done! You guessed my number!")
  exit(0)
````
